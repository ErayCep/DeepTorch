```python
import torch
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 18})
```

```python
x = torch.linspace(-3, 3, 101)

def NNoutputx(act_func):
    act_func = getattr(torch, act_func)
    return act_func(x)
```

```python
activation_funcs = ['relu', 'sigmoid', 'tanh']

fig = plt.figure(figsize=(10,8))

for act_func in activation_funcs:
    plt.plot(x, NNoutputx(act_func), label=act_func, linewidth=3)
    
dashlinecol = [.7, .7, .7]
plt.plot(x[[0,-1]], [0,0], '--', color=dashlinecol)
plt.plot(x[[0,-1]], [1,1], '--', color=dashlinecol)
plt.plot([0,0], [-1,3], '--', color=dashlinecol)

plt.legend()
plt.show()
```

```python
def NNoutput(act_func):
    act_func = getattr(torch.nn, act_func)
    return act_func()
```

```python
activation_funcs = ['ReLU6', 'Hardshrink', 'LeakyReLU']

fig = plt.figure(figsize=(10,8))

for act_func in activation_funcs:
    plt.plot(x, NNoutput(act_func)(x), label=act_func, linewidth=3)

dashlinecol = [.7, .7, .7]
plt.plot(x[[0,-1]], [0,0], '--', color=dashlinecol)
plt.plot(x[[0,-1]], [1,1], '--', color=dashlinecol)
plt.plot([0,0], [-1,3], '--', color=dashlinecol)

plt.legend()
plt.show()
```
